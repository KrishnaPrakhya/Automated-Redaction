```python
import cv2
import numpy as np
import pytesseract

def invert_image(image):
    return cv2.bitwise_not(image)

def grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def apply_threshold(image, threshold_value=210):
    _, im_bw = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)
    return im_bw

def thin_font(image):
    image = cv2.bitwise_not(image)
    kernel = np.ones((1, 1), np.uint8)
    image = cv2.erode(image, kernel, iterations=1)
    return cv2.bitwise_not(image)

def thick_font(image):
    image = cv2.bitwise_not(image)
    kernel = np.ones((2, 2), np.uint8)
    image = cv2.dilate(image, kernel, iterations=1)
    return cv2.bitwise_not(image)

def getSkewAngle(cvImage) -> float:
    newImage = cvImage.copy()
    gray = cv2.cvtColor(newImage, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (9, 9), 0)
    thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (30, 5))
    dilate = cv2.dilate(thresh, kernel, iterations=2)
    contours, _ = cv2.findContours(dilate, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    contours = sorted(contours, key=cv2.contourArea, reverse=True)

    largestContour = contours[0] if contours else None
    if largestContour is None:
        return 0  
    
    minAreaRect = cv2.minAreaRect(largestContour)
    angle = minAreaRect[-1]

    if angle < -45:
        angle = 90 + angle

    if abs(angle) > 89: 
        angle = 0  

    return -1.0 * angle

def rotateImage(cvImage, angle: float):
    (h, w) = cvImage.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    return cv2.warpAffine(cvImage, M, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)

def deskew(image):
    angle = getSkewAngle(image)
    print(angle)
    if abs(angle) > 0.5:  
        return rotateImage(image, -angle)
    return image

def extract_text_from_roi(image, roi_coordinates):
    x, y, w, h = roi_coordinates
    roi = image[y:y+h, x:x+w]
    text = pytesseract.image_to_string(roi)
    return text, roi

def remove_lines(image):
    horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (40, 1))
    detect_horizontal = cv2.morphologyEx(image, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

    vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 40))
    detect_vertical = cv2.morphologyEx(image, cv2.MORPH_OPEN, vertical_kernel, iterations=2)

    return detect_horizontal + detect_vertical

def remove_noise(image):
    kernel = np.ones((1, 1), np.uint8)
    return cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

def sharpen_image(image):
    kernel = np.array([[-1, -1, -1], [-1, 9,-1], [-1, -1, -1]])
    return cv2.filter2D(image, -1, kernel)

def contrast_stretch(image):
    xp = [0, 64, 128, 192, 255]
    fp = [0, 16, 128, 240, 255]
    table = np.interp(np.arange(0, 256), xp, fp).astype('uint8')
    return cv2.LUT(image, table)

def adaptive_threshold(image):
    return cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 10)

def adjust_contrast(image, alpha=1.5, beta=0):
    return cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

def main_body(image):
    base_img = image.copy()
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (7, 7), 0)

    thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_OTSU + cv2.THRESH_BINARY_INV)[1]
    cv2.imwrite("thresh_output.jpg", thresh)

    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 25))
    dilate = cv2.dilate(thresh, kernel, iterations=1)
    cv2.imwrite("dilated_output.jpg", dilate)

    cnts, _ = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = sorted(cnts, key=lambda x: cv2.boundingRect(x)[1])

    for c in cnts:
        x, y, w, h = cv2.boundingRect(c)
        if h > 200 and w > 250:  
            roi = base_img[y:y+h, x:x+w]
            text = pytesseract.image_to_string(roi)
            print(f"Detected text in ROI:\n{text}")
            cv2.imwrite(f"roi_{x}_{y}.jpg", roi)

    full_image_text = pytesseract.image_to_string(image)
    print(f"Detected text in full image:\n{full_image_text}")

def process_image(image_file):
    img = cv2.imread(image_file)
    
    img = remove_noise(img)
    
#     img = sharpen_image(img)
    
    deskewed_img = deskew(img)
    cv2.imwrite("deskewed_image.jpg", deskewed_img)
    print("1:::::",pytesseract.image_to_string(deskewed_img))
    
    inverted_img = invert_image(deskewed_img)
    cv2.imwrite("inverted_image.jpg", inverted_img)
    print("2:::::",pytesseract.image_to_string(inverted_img))
    
    
    contrast = contrast_stretch(deskewed_img)
    cv2.imwrite("contrast.jpg", contrast)
    print("3:::::",pytesseract.image_to_string(contrast))
    
    gray_img = grayscale(deskewed_img)
    cv2.imwrite("grayscale_image.jpg", gray_img)
    print("4:::::",pytesseract.image_to_string(gray_img))
    
    bw_img = adaptive_threshold(gray_img)
    cv2.imwrite("bw_image.jpg", bw_img)
    print("5:::::BW",pytesseract.image_to_string(bw_img))
    main_body(img)
    # Step 8: Remove lines and enhance font thickness
#     processed_img = thick_font(bw_img)
#     cv2.imwrite("final_processed_image.jpg", processed_img)
    
    full_text = pytesseract.image_to_string(deskewed_img)
    print(f"Detected text from deskewed image:\n{full_text}")
    
#     roi_coordinates = (x, y, w, h)  # Define ROI coordinates
#     text, roi = extract_text_from_roi(deskewed_img, roi_coordinates)
    
    return deskewed_img




image_file = "12.jpg"
processed_image = process_image(image_file)

```

```python
from pytesseract import Output
image_file = "12.jpg"
def detect_text_properties(image):
    
    data = pytesseract.image_to_data(image, output_type=Output.DICT)
    text_info = []
    
    for i in range(len(data['text'])):
        if data['text'][i].strip():  
            x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
            text = data['text'][i]
            font_size = h  
            text_info.append({
                "text": text,
                "position": (x, y, w, h),
                "font_size": font_size
            })
    
    return text_info
```

```python
def redact_text(image, text_info, redact_list, replacement_text="REDACTED"):
   
    
    redacted_img = image.copy()
    
    for info in text_info:
        x, y, w, h = info["position"]
        text = info["text"]
        font_size = info["font_size"]
        
        if text in redact_list:
            cv2.rectangle(redacted_img, (x, y), (x+w, y+h), (1, 1, 1), -1)
            
            font = cv2.FONT_HERSHEY_SIMPLEX
            font_scale = font_size / 30  
            thickness = 1
            
            (text_width, text_height), baseline = cv2.getTextSize(replacement_text, font, font_scale, thickness)
            
            while text_width > w or text_height > h and font_scale>0.5:
                font_scale-=0.1
                (text_width, text_height), baseline = cv2.getTextSize(replacement_text, font, font_scale, thickness)
                
                
            text_x = x + (w - text_width) // 2
            text_y = y + (h + text_height) // 2 - baseline
            
            cv2.putText(redacted_img, replacement_text, (text_x, text_y), font, font_scale, (255, 255, 255), thickness)
    
    return redacted_img
```

```python
import cv2

def redact_faces(image_path, redact_type='black'):
    image = cv2.imread(image_path)
    
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for (x, y, w, h) in faces:
        if redact_type == 'black':
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 0), -1)
        elif redact_type == 'blur':
            face = image[y:y + h, x:x + w]
            blurred_face = cv2.GaussianBlur(face, (99, 99), 30)
            image[y:y + h, x:x + w] = blurred_face
        elif redact_type == 'pixelate':
            face = image[y:y + h, x:x + w]
            pixelated_face = cv2.resize(face, (16, 16), interpolation=cv2.INTER_LINEAR)
            pixelated_face = cv2.resize(pixelated_face, (w, h), interpolation=cv2.INTER_NEAREST)
            image[y:y + h, x:x + w] = pixelated_face
    
    return image

redacted_image = redact_faces('redacted_image.jpg', redact_type='blur')

cv2.imwrite('redacted_image.jpg', redacted_image)
cv2.imshow('Redacted Image', redacted_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
